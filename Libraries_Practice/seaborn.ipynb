```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker",style="smoker", size="size",
)
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True, bins= 15)
```

```python
sns.catplot(data=tips, kind="swarm", x="day", y="total_bill", hue="smoker")
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", hue="time")
```

```python
sns.pairplot(data=tips, hue="smoker")
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Create a barplot by using matplotlib
sns.barplot(data=tips, x="sex", y="total_bill")
# Add title and labels to the plot
plt.title("Barplot")
plt.xlabel("Gender")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Create a boxplot by using matplotlib
sns.boxplot(data=tips, x="sex", y="total_bill")
# Add title and labels to the plot
plt.title("Boxplot")
plt.xlabel("Gender")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Create a violinplot by using matplotlib
sns.violinplot(data=tips, x="sex", y="total_bill")
# Add title and labels to the plot
plt.title("Violinplot")
plt.xlabel("Gender")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Create a histplot by using matplotlib
sns.histplot(data=tips, x="total_bill")
# Add title and labels to the plot
plt.title("Histogram")
plt.xlabel("Total Bill")
plt.ylabel("Count")
plt.show()
```

```python
# Create a distplot by using seaborn
sns.displot(tips["total_bill"])
# Add title and labels to the plot
plt.title("Distribution Plot")
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.show()
```

```python
# Create a rugplot by using seaborn
sns.rugplot(tips["total_bill"])
# Add title and labels to the plot
plt.title("Rug Plot")
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.show()
```

```python
# create a kde plot by using seaborn
sns.kdeplot(tips["total_bill"])
# Add title and labels to the plot
plt.title("KDE Plot")
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.show()
```

```python
# Create a factor plot by using seaborn
sns.stripplot(data=tips, x="day", y="total_bill")
# Add title and labels to the plot
plt.title("Stripplot")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
# Create a swarm plot by using seaborn
sns.swarmplot(data=tips, x="day", y="total_bill")
# Add title and labels to the plot
plt.title("Swarmplot")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python

# Remove non-numeric columns if they are not needed for correlation analysis
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix of the numeric DataFrame
corr_matrix = numeric_df.corr()

# Create the heatmap
sns.heatmap(corr_matrix, annot=True)

# Add title and labels to the plot
plt.title("Heatmap")
plt.xlabel("Features")
plt.ylabel("Features")

# Display the heatmap
plt.show()
```
